```python
#Employee 

class Employee:

  Empdetails = dict()

  def __init__(self,fname,lname,gender, empid,edu, dept, bankact, title):
    self.firstName = fname
    self.lastName = lname
    self.empid = empid
    self.education = edu
    self.gender = gender
    self.department = dept
    self.bankacct = bankact
    self.title = title
    Employee.Empdetails[empid] = self

  def add():
    Empl = input("Enter the employee's fname,lname,gender, empid,edu, dept, bankact, title: ").split(',')
    
    Empdetail = Employee(*Empl)
    
  def remove():

    empid = input("Enter the empid of the details you want to remove:")

    if empid not in Employee.Empdetails:
      print("record does not exist")
    else:
      del Employee.Empdetails[empid]

  def setattribute(): 

     var = input("Enter the empid of the employee you want to change and the attribute you want to change. Please separate them with a comma:").split(",")
    
     empid, others = var

     if others in ['firstName','lastName','gender', 'empid','education', 'department', 'bankacc']:

       varnew = input("Enter the new attribute value")
       
       setattr(Employee.Empdetails[empid], others, varnew)
```

```python
def empOptions(empType):
  print('''
             press a: to add new employee details
             press r: to remove employee details
             press m: to modify the employees details''')
  
  
  choices = input("enter an option of your choice: ") 
  if choices == "a":
    if empType==1:
      EmpFullTime.add()
    elif empType==2:
      EmpPartTime.add()
    print("You have successfully added your Employee's detail ")
  
  elif choices == "r":
    if empType==1:
      EmpFullTime.remove()
    elif empType==2:
      EmpPartTime.remove()
   
    print("You have successfully removed the Employee's information from the database: ")

  elif choices == "m":
    if empType==1:
      EmpFullTime.setattribute()
    elif empType==2:
      EmpPartTime.setattribute()
    
    print("Employee's info has been updated ")

  else:
    print("Oop!, You've just entered an invalid option. Try Again ")
```

```python
def salaryOptions(emptype):
  print('''
             press 1: to create salary info
             press 2: to display gross salary
             press 3: to add bonus
             press 4: to add overtime
             press 5: to change tax percentage
             press 6: to change insurance value
             press 7: to deduct penalty fee
             press 8: to get net salary
             
            ''')
  choices = int(input("enter an option of your choice "))
  if choices == 1:
    Salary.CreateSalaryinfo()
  else:
    empid=  input('Insert empid ')   
    if choices == 2:
      Salary.displayGross_Salary(empid)
    elif choices == 3:
      Salary.AddBonus(empid)
    elif choices == 4:
      Salary.AddOvertime(empid)
    elif choices == 5:
      Salary.Get_Tax_percent(empid)
    elif choices == 6:
      Salary.ChangeInsuranceValue(empid,emptype)
    elif choices == 7:
      Salary.DeductPenaltyFee(empid)
    elif choices == 8:
      Salary.GetNetSalary(empid,emptype)
    else:
      print("Sorry, invalid option")
```

```python
#salary class

class Salary:
  partTimehourlyRate=20
  salary = dict() 

  def __init__(self,gross_salary,empType,empid):
    self.tax_percent = 0
    self.gross_salary = int(gross_salary)                    
    self.insurance = 0           
    self.overtime = 0
    self.bonus = 0
    self.penalty_fee = 0
    self.empid=empid
    self.empType=empType
    Salary.salary[empid]=self
    self.empHoursPart=0

  def CreateSalaryinfo():
    salarydata=input('''For part time insert gross salary as 0 \n
                                      enter employee's gross_salary,empType,empid: ''').split(',')
    #####
    salary_=Salary(*salarydata)
    if salarydata[-2] ==1:
      Salary.salary[salarydata[-1]] = salary_
    elif salarydata[-2] ==1: 
      Salary.salary[salarydata[-1]] = salary_
      Salary.salary[salarydata[-1]].empHoursPart = int(input('Insert the number of part time working hours as integer: ').split())
      Salary.salary[salarydata[-1]].gross_salary=Salary.salary[salarydata[-1]].empHoursPart * Salary.partTimehourlyRate

  def displayGross_Salary(empid): 
      print(Salary.salary[empid].gross_salary) 

  def AddBonus(empid):  
    Salary.salary[empid].bonus=int(input("Enter your bonus: "))

   
  def AddOvertime(empid):
    Salary.salary[empid].overtime = int(input("Enter your overtime value: ")) #overtime changed to str
    
  def Get_Tax_percent(empid):
    if  Salary.salary[empid].gross_salary + Salary.salary[empid].bonus <= 10000:
      print("Congratulations,You are tax free. Work Hard to earn more ")
    elif (Salary.salary[empid].gross_salary + Salary.salary[empid].bonus < 100000):
      Salary.salary[empid].tax_percent = 10 
      print(f"You are required to pay {Salary.salary[empid].tax_percent}% of your income ")
    else:
      Salary.salary[empid].tax_percent = 25 
      print(f"You are required to pay {Salary.salary[empid].tax_percent}% of your income ")

  def ChangeInsuranceValue(empid,emptype):
    if emptype==1:
      Salary.salary[empid].insurance = int(input("Enter your insurance value "))
      print(f"your New insurance value is: {Salary.salary[empid].insurance } ")

  def DeductPenaltyFee(empid):
    Salary.salary[empid].penalty_fee =int(input('Enter your penalty fee: '))

  def GetNetSalary(empid,emptype):

    if emptype== 1:
      Tax = Salary.salary[empid].tax_percent/100* (Salary.salary[empid].gross_salary + Salary.salary[empid].bonus + Salary.salary[empid].overtime - Salary.salary[empid].insurance-Salary.salary[empid].penalty_fee)
    elif emptype== 2:
      Tax = Salary.salary[empid].tax_percent/100* (Salary.salary[empid].gross_salary + Salary.salary[empid].bonus + Salary.salary[empid].overtime -Salary.salary[empid].penalty_fee)
    NetSalary =  - Tax+(Salary.salary[empid].gross_salary + Salary.salary[empid].bonus + Salary.salary[empid].overtime - Salary.salary[empid].insurance-Salary.salary[empid].penalty_fee)
    print(NetSalary)
```
